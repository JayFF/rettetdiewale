```python
import pandas as pd
import re
```

```python
data = pd.read_csv("btw17_kerg.csv",";")
```

```python
data2 = data.drop([column for column in list(data) if (re.search("Unnamed: .*", column) and "Zweitstimmen" not in list(data[column]))], axis=1)
```

```python
data2.drop(data2.index[1], inplace=True)
```

```python
data2;
```

```python
for index, column in enumerate(list(data2)):
    print(index, column)
```

```python
for index, column in enumerate(list(data2)):
    if re.match("Unnamed: .*", column):
        #print((str(list(data)[index])+" Zweitstimmen"+str(index)))
        data2 = data2.rename(columns={column: (str(list(data2)[index-1])+" Zweitstimmen")})
```
